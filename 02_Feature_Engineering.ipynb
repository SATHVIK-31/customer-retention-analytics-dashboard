```python
df['Review Rating'] = df['Review Rating'].fillna(
    df['Review Rating'].median()
)
```

```python
df['Frequency of Purchases'] = df['Frequency of Purchases'].replace({
    'Bi-Weekly':'Fortnightly',
    'Every 3 Months':'Quarterly'
})
```

```python
frequency_map = {
    'Weekly':5,
    'Fortnightly':4,
    'Monthly':3,
    'Quarterly':2,
    'Annually':1
}

df['Frequency Score'] = df['Frequency of Purchases'].map(frequency_map)
```

```python
df['Subscription Score'] = np.where(
    df['Subscription Status'] == 'Yes',
    1,
    0
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Previous Purchases_N',
    'Frequency Score_N',
    'Review Rating_N']] = scaler.fit_transform(
        df[['Previous Purchases',
            'Frequency Score',
            'Review Rating']]
)
```

```python
df[['Previous Purchases_N',
    'Frequency Score_N',
    'Review Rating_N']].head()
```

```python
df['Subscription Score'] = np.where(
    df['Subscription Status'] == 'Yes',
    1,
    0
)
```

```python
df['Loyalty Score V1'] = (
    0.4 * df['Previous Purchases_N']
    + 0.3 * df['Frequency Score_N']
    + 0.2 * df['Subscription Score']
    + 0.1 * df['Review Rating_N']
)
```

```python
df['Loyalty Band'] = pd.qcut(
    df['Loyalty Score V1'],
    q=4,
    labels=['Low','Medium','High','Very High']
)
```

```python
df['Loyalty Band'].value_counts()
```

```python
df[['Purchase Amount_N']] = scaler.fit_transform(
    df[['Purchase Amount (USD)']]
)
```

```python
df[['Purchase Amount (USD)','Purchase Amount_N']].head()
```

```python
df['Loyalty Score V2'] = (
    0.5 * df['Previous Purchases_N']
    + 0.3 * df['Purchase Amount_N']
    + 0.2 * df['Review Rating_N']
)
```

```python
df[['Loyalty Score V1',
    'Loyalty Score V2']].describe()
```

```python
df['Promo Dependency Score'] = (
    (df['Discount Applied'] == 'Yes').astype(int)
    +
    (df['Promo Code Used'] == 'Yes').astype(int)
)
```

```python
def promo_segment(x):
    if x == 2:
        return 'High'
    elif x == 1:
        return 'Medium'
    else:
        return 'Low'

df['Promo Dependency Level'] = (
    df['Promo Dependency Score']
    .apply(promo_segment)
)
```

```python
df['Promo Dependency Level'].value_counts()
```

```python
df[['Loyalty Score V1',
    'Loyalty Score V2']].describe()
```

```python
df['Promo Dependency Level'].value_counts()
```

```python
df['Customer Value Score'] = (
    df['Purchase Amount (USD)']
    *
    df['Previous Purchases']
)
```

```python
df[['Purchase Amount (USD)',
    'Previous Purchases',
    'Customer Value Score']].head()
```

```python
df['Value Tier'] = pd.qcut(
    df['Customer Value Score'],
    q=4,
    labels=[
        'Bronze',
        'Silver',
        'Gold',
        'Platinum'
    ]
)
```

```python
df['Value Tier'].value_counts()
```

```python
def customer_segment(row):

    if (
        row['Value Tier'] == 'Platinum'
        and
        row['Promo Dependency Level'] == 'Low'
    ):
        return 'VIP Loyalist'

    elif (
        row['Promo Dependency Level'] == 'High'
    ):
        return 'Bargain Hunter'

    elif (
        row['Value Tier'] in ['Gold', 'Silver']
    ):
        return 'Growth Customer'

    else:
        return 'Low Engagement'


df['Customer Segment'] = df.apply(
    customer_segment,
    axis=1
)
```

```python
df['Customer Segment'].value_counts()
```

```python
df.columns.tolist()
```

```python
df.to_csv(
    '../processed/customer_features.csv',
    index=False
)
```
