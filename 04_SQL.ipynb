```python
import sqlite3
import pandas as pd

df = pd.read_csv('../processed/customer_features.csv')

conn = sqlite3.connect('../sql/customer_retention.db')

df.to_sql(
    'customer_features',
    conn,
    if_exists='replace',
    index=False
)

print("Database Created Successfully")
```

```python
query = """
PRAGMA table_info(customer_features);
"""

pd.read_sql_query(query, conn)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../sql/customer_retention.db')

query = """
SELECT "Customer Segment",
       COUNT(*) AS Customer_Count
FROM customer_features
GROUP BY "Customer Segment"
"""

result = pd.read_sql_query(query, conn)

result
```

```python
query = """
SELECT "Value Tier",
       COUNT(*) AS Customer_Count
FROM customer_features
GROUP BY "Value Tier"
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT "Promo Dependency Level",
       COUNT(*) AS Customer_Count
FROM customer_features
GROUP BY "Promo Dependency Level"
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT Category,
       SUM("Purchase Amount (USD)") AS Revenue
FROM customer_features
GROUP BY Category
ORDER BY Revenue DESC
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT "Customer Segment",
       ROUND(AVG("Purchase Amount (USD)"),2) AS Avg_Spend
FROM customer_features
GROUP BY "Customer Segment"
ORDER BY Avg_Spend DESC
"""

pd.read_sql_query(query, conn)
```
